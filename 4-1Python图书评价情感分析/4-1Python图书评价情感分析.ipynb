## 图书评价情感分析

任务1 定义函数get_lst_from_file  
功能：读取文本文件，读取每一行添加到列表中  
输入参数：文件名  
返回值：列表  
提示：文本文件中有中文，编码格式未知，可能是'gbk', 'utf-8', 'utf-16', 'utf-8-sig'，需要多次试验，读取文件时，需要加上参数encoding指定格式

```python
#读取文件，尝试常见的编码格式
def get_lst_from_file(file_path):
    lst_encoding = ['gbk', 'utf-8', 'utf-16', 'utf-8-sig']
    i = 0
    f = None
    lst = []
    while i < len(lst_encoding):
        try:
            f = open(file_path, mode='r', encoding = lst_encoding[i])
            lst.extend(f.readlines())
            break
        except:
            i += 1
        finally:
            if f and not f.closed:
                f.close()
    return lst

get_lst_from_file("data/负面评价词库.txt")
```

任务2 定义函数get_word_lst   
功能：调用get_lst_from_file函数，对返回的列表进行处理，包括去掉每个元素头和尾的空格、回车等字符文件名  
输入参数：文件名  
返回值： 列表  

```python
def get_word_lst(file_path):
    word_list = [line.strip() for line in get_lst_from_file(file_path)]
    new_list = list(filter(None, word_list))
    return new_list

get_word_lst("data/负面评价词库.txt")
```

任务3 定义函数check_comment  
功能：检查一条书评是正面的、中性的还是负面的  
输入参数：一条书评，正面评价词库列表，负面评价词库列表    
返回值：正数代表正面评价，0代表中性评价，负数代表正面评价  
提示：试试用正则表达式解决，可以简化程序，并提高匹配效率

```python
#检查一条书评是否是正面的
def check_comment(comment, positive_words, negative_words):
    import re
    positive_comment = len(re.findall('|'.join(positive_words), comment))
    negative_comment = len(re.findall('|'.join(negative_words), comment))
    return  positive_comment - negative_comment

check_comment("挺好，给别人买的，反映不错。", get_word_lst("data/正面评价词库.txt"), get_word_lst("data/负面评价词库.txt"))
```

任务4 定义函数evaluate_book  
功能：调用check_comment函数，对每条评价进行分析  
输入参数：书评列表，正面评价词库列表，负面评价词库列表  
返回值：返回一个元组，元素分别是正面评价数量、中性评价数量、负数评价数量   
提示：试试用正则表达式解决，可以简化程序，并提高匹配效率

```python
def evaluate_book(book_comments, positive_words, negative_words):
    positive = 0
    neutral = 0
    negative = 0
    for comment in book_comments:
        comment_type = check_comment(comment, positive_words, negative_words)
        if comment_type > 0:
            positive += 1
            print("正面评价：" + comment)
        elif comment_type == 0:
            neutral += 1
            print("中性评价：" + comment)
        else:
            negative += 1
            print("负面评价：" + comment)
    return positive, neutral, negative

evaluate_book(get_word_lst("data/书评.txt"), get_word_lst("data/正面评价词库.txt"), get_word_lst("data/负面评价词库.txt"))
```

开始编写主程序


任务5 分别调用函数get_word_lst，得到书评列表、正面评价词库列表、负面评价词库列表，调用evaluate_book函数，输出评价结果

```python
book_comments =  get_word_lst('data/书评.txt')
positive_words = get_word_lst('data/正面评价词库.txt')
negative_words = get_word_lst('data/负面评价词库.txt')

print("正面评价{0}条，中性评价{1}条，负面评价{2}条".format(* evaluate_book(book_comments, positive_words, negative_words)))
```
